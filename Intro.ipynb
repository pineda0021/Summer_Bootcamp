# **Python vs R: What's the Difference?**
---
## The main difference is that **Python is a general-purpose programming** language, while R has its roots in statistical analysis.

## R offers superior support and extensive libraries because it was created for data statistics. Python works much better when creating and deploying applications. However, R and its libraries implement a wide range of statistical and graphical approaches for data analysis.

# **What is GitHub?**
---
## GitHub is a web-based platform for **version control** and collaborative software development. It allows developers to host, review, manage, and collaborate on code project. Also, It provides tools for version tracking, bug tracking, and feature requests.

•	Repositories

•	Pull requests

#**How to create a Github account**

1. **Go to the GitHub website:**

   Open your web browser and go to [GitHub](https://github.com).

   Video on how to create the GitHub: [https://www.youtube.com/watch?v=Gn3w1UvTx0A](https://www.youtube.com/watch?v=Gn3w1UvTx0A)

2. **Sign Up:**

   Click on the "Sign up" button located at the top right corner of the page.

3. **Enter Your Information:**

   Fill in the required details, such as your email address, a username, and a password.
   Click "Create account."

4. **Verify Your Account:**

   You will receive a verification email. Open it and click on the link to verify your email address.

5. **Set Up Your Profile:**

   Follow the prompts to set up your profile. You can skip optional steps if you prefer.

6. **Confirm Account Creation:**

   Once your account is set up, you should see your GitHub dashboard.
